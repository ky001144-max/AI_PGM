```python
class Shape:
    def __init__(self, name):
        self.name = name
    def area(self):
        raise NotImplementedError("Subclasses must implement this method")

class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius
    def area(self):
        return 3.14159 * self.radius * self.radius
    
class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__("Rectangle")
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    
shapeList = [Circle(5), Rectangle(4, 6),]
for shape in shapeList:
    print(f"{shape.name} area: {shape.area()}")
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        raise ZeroDivisionError("0으로 나눌 수 없습니다.")
    return x / y

def calculator():
    print("=" * 30)
    print("   파이썬 CLI 계산기 프로그램   ")
    print("=" * 30)

    while True:
        print("\n[원하는 연산을 선택하세요]")
        print("1: 더하기 (+)")
        print("2: 빼기 (-)")
        print("3: 곱하기 (*)")
        print("4: 나누기 (/)")
        print("q: 프로그램 종료")
        
        choice = input("선택 (1/2/3/4/q): ").strip()

        # 종료 조건 확인
        if choice.lower() == 'q':
            print("\n계산기를 종료합니다. 이용해 주셔서 감사합니다!")
            break

        # 올바른 메뉴를 선택했는지 확인
        if choice not in ['1', '2', '3', '4']:
            print("❌ 잘못된 입력입니다. 1, 2, 3, 4 또는 q를 입력해주세요.")
            continue

        # 숫자 입력 받기 (예외 처리 포함)
        try:
            num1 = float(input("첫 번째 숫자를 입력하세요: "))
            num2 = float(input("두 번째 숫자를 입력하세요: "))
        except ValueError:
            print("❌ 잘못된 입력입니다. 숫자만 입력할 수 있습니다.")
            continue

        # 연산 수행 및 결과 출력
        try:
            if choice == '1':
                result = add(num1, num2)
                print(f"▶ 결과: {num1} + {num2} = {result}")
            elif choice == '2':
                result = subtract(num1, num2)
                print(f"▶ 결과: {num1} - {num2} = {result}")
            elif choice == '3':
                result = multiply(num1, num2)
                print(f"▶ 결과: {num1} * {num2} = {result}")
            elif choice == '4':
                result = divide(num1, num2)
                print(f"▶ 결과: {num1} / {num2} = {result}")
        except ZeroDivisionError as e:
            print(f"❌ 오류: {e}")
            
        print("-" * 30)

if __name__ == "__main__":
    calculator()
```

```python
import tkinter as tk
from tkinter import messagebox

# 전역 변수로 계산기 식을 저장할 변수 선언
expression = ""

def press(num):
    """숫자나 연산자 버튼을 눌렀을 때 실행되는 함수"""
    global expression
    expression = expression + str(num)
    equation.set(expression)

def equalpress():
    """= 버튼을 눌러 결과를 계산하는 함수"""
    global expression
    try:
        # eval 함수를 사용해 문자열로 된 수식을 계산
        # 예: "2+3*5" -> 17
        total = str(eval(expression))
        equation.set(total)
        expression = total  # 계산 결과를 이어서 사용할 수 있도록 저장
    except ZeroDivisionError:
        messagebox.showerror("오류", "0으로 나눌 수 없습니다.")
        clear()
    except Exception:
        messagebox.showerror("오류", "수식이 올바르지 않습니다.")
        clear()

def clear():
    """C 버튼을 눌러 입력을 초기화하는 함수"""
    global expression
    expression = ""
    equation.set("")

# --- GUI 창 설정 ---
root = tk.Tk()
root.title("파이썬 GUI 계산기")
root.geometry("300x400")
root.configure(bg="#f0f0f0")

# 텍스트를 화면에 표시하기 위한 tkinter 변수
equation = tk.StringVar()

# 1. 디스플레이 창 (결과창)
display_frame = tk.Frame(root, width=300, height=50, bd=0, highlightbackground="#cccccc", highlightcolor="#cccccc", highlightthickness=1)
display_frame.pack(side=tk.TOP, fill=tk.BOTH, expand=True)

display_field = tk.Entry(display_frame, textvariable=equation, font=('Arial', 20, 'bold'), fg="#333333", bg="#ffffff", bd=0, justify=tk.RIGHT)
display_field.pack(fill=tk.BOTH, ipadx=8, ipady=15, expand=True)

# 2. 버튼 프레임
buttons_frame = tk.Frame(root, bg="#f0f0f0")
buttons_frame.pack(side=tk.TOP, fill=tk.BOTH, expand=True)

# 버튼 설정 레이아웃 (텍스트, 행, 열, 배경색)
button_config = [
    ('7', 1, 0, '#ffffff'), ('8', 1, 1, '#ffffff'), ('9', 1, 2, '#ffffff'), ('/', 1, 3, '#f9a825'),
    ('4', 2, 0, '#ffffff'), ('5', 2, 1, '#ffffff'), ('6', 2, 2, '#ffffff'), ('*', 2, 3, '#f9a825'),
    ('1', 3, 0, '#ffffff'), ('2', 3, 1, '#ffffff'), ('3', 3, 2, '#ffffff'), ('-', 3, 3, '#f9a825'),
    ('C', 4, 0, '#e57373'), ('0', 4, 1, '#ffffff'), ('=', 4, 2, '#4caf50'), ('+', 4, 3, '#f9a825'),
]

# 버튼 생성 및 배치
for (text, row, col, bg_color) in button_config:
    # 람다(lambda)를 사용해 각 버튼 클릭 시 올바른 인자 전달
    if text == '=':
        action = equalpress
        fg_color = '#ffffff'
    elif text == 'C':
        action = clear
        fg_color = '#ffffff'
    else:
        action = lambda x=text: press(x)
        fg_color = '#333333' if bg_color == '#ffffff' else '#ffffff'

    btn = tk.Button(
        buttons_frame, text=text, font=('Arial', 14, 'bold'), 
        bg=bg_color, fg=fg_color, bd=1, relief=tk.RAISED, command=action
    )
    # 버튼이 프레임 꽉 차게 그리드 배치
    btn.grid(row=row, column=col, sticky="nsew", padx=2, pady=2)

# 그리드 비율 설정 (버튼이 창 크기에 맞게 조절되도록 함)
for i in range(4):
    buttons_frame.columnconfigure(i, weight=1)
for i in range(1, 5):
    buttons_frame.rowconfigure(i, weight=1)

# 프로그램 실행
root.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from tkinter import ttk
import tkinter.font as tkfont
import os

class Notepad:
    def __init__(self, root):
        self.root = root
        self.root.title("제목 없음 - 파이썬 메모장")
        self.root.geometry("600x500")
        
        self.file_path = None
        
        # 기본 폰트 정보 (처음 실행 시 기준)
        self.current_font_family = "맑은 고딕"
        self.current_font_size = 12
        self.is_bold = False
        self.is_italic = False
        self.is_underline = False
        self.is_strike = False

        # 1. 텍스트 창 생성
        self.text_area = tk.Text(self.root, undo=True)
        self.text_area.pack(fill=tk.BOTH, expand=True)

        # 초기 기본 폰트 적용
        self.update_base_font()

        # 2. 스크롤바 추가
        self.scroll_bar = tk.Scrollbar(self.text_area)
        self.scroll_bar.pack(side=tk.RIGHT, fill=tk.Y)
        self.text_area.config(yscrollcommand=self.scroll_bar.set)
        self.scroll_bar.config(command=self.text_area.yview)

        # 3. 상단 메뉴바 구성
        self.menu_bar = tk.Menu(self.root)
        self.root.config(menu=self.menu_bar)

        # [파일 메뉴]
        self.file_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.menu_bar.add_cascade(label="파일(F)", menu=self.file_menu)
        self.file_menu.add_command(label="새로 만들기(N)", command=self.new_file, accelerator="Ctrl+N")
        self.file_menu.add_command(label="열기(O)...", command=self.open_file, accelerator="Ctrl+O")
        self.file_menu.add_command(label="저장(S)", command=self.save_file, accelerator="Ctrl+S")
        self.file_menu.add_command(label="다른 이름으로 저장(A)...", command=self.save_as_file)
        self.file_menu.add_separator()
        self.file_menu.add_command(label="끝내기(X)", command=self.quit_application)

        # [서식 메뉴]
        self.format_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.menu_bar.add_cascade(label="서식(O)", menu=self.format_menu)
        self.format_menu.add_command(label="글꼴/효과 설정...", command=self.open_font_window)

        # 단축키 바인딩
        self.root.bind("<Control-n>", lambda event: self.new_file())
        self.root.bind("<Control-o>", lambda event: self.open_file())
        self.root.bind("<Control-s>", lambda event: self.save_file())

        self.tag_id = 0

    def update_base_font(self):
        """본문의 기본 폰트 서식을 설정합니다. (태그가 지정되지 않은 영역에 적용)"""
        font_style = tkfont.Font(
            family=self.current_font_family,
            size=self.current_font_size,
            weight="bold" if self.is_bold else "normal",
            slant="italic" if self.is_italic else "roman",
            underline=1 if self.is_underline else 0,
            overstrike=1 if self.is_strike else 0
        )
        self.text_area.config(font=font_style)

    def open_font_window(self):
        """글꼴, 크기 및 효과를 변경하는 설정 창"""
        has_selection = True
        try:
            # 블록 지정된 영역이 있는지 확인
            start_idx = self.text_area.index("sel.first")
            end_idx = self.text_area.index("sel.last")
        except tk.TclError:
            # 선택 영역이 없으면 현재 커서 위치를 기준으로 잡음
            has_selection = False
            insert_idx = self.text_area.index("insert")

        font_window = tk.Toplevel(self.root)
        font_window.title("글꼴 및 효과 설정")
        font_window.geometry("380x300")
        font_window.resizable(False, False)
        
        font_window.transient(self.root)
        font_window.grab_set()

        frame = tk.Frame(font_window, padx=20, pady=15)
        frame.pack(fill=tk.BOTH, expand=True)

        # 1. 글꼴 선택
        tk.Label(frame, text="글꼴:").grid(row=0, column=0, sticky="w", pady=5)
        filtered_fonts = [f for f in sorted(list(tkfont.families())) if not f.startswith('@')]
        font_cb = ttk.Combobox(frame, values=filtered_fonts, width=22, state="readonly")
        font_cb.set(self.current_font_family)
        font_cb.grid(row=0, column=1, pady=5, padx=5, columnspan=2)

        # 2. 글자 크기 선택
        tk.Label(frame, text="크기:").grid(row=1, column=0, sticky="w", pady=5)
        sizes = [8, 9, 10, 11, 12, 14, 16, 18, 20, 22, 24, 28, 32, 36, 48, 72]
        size_cb = ttk.Combobox(frame, values=sizes, width=22, state="readonly")
        size_cb.set(self.current_font_size)
        size_cb.grid(row=1, column=1, pady=5, padx=5, columnspan=2)

        # 3. 효과 선택 (체크박스)
        tk.Label(frame, text="효과:").grid(row=2, column=0, sticky="nw", pady=10)
        
        bold_var = tk.BooleanVar(value=self.is_bold if not has_selection else False)
        italic_var = tk.BooleanVar(value=self.is_italic if not has_selection else False)
        underline_var = tk.BooleanVar(value=self.is_underline if not has_selection else False)
        strike_var = tk.BooleanVar(value=self.is_strike if not has_selection else False)

        tk.Checkbutton(frame, text="굵게 (Bold)", variable=bold_var).grid(row=2, column=1, sticky="w", pady=2, padx=5)
        tk.Checkbutton(frame, text="기울임 (Italic)", variable=italic_var).grid(row=2, column=2, sticky="w", pady=2, padx=5)
        tk.Checkbutton(frame, text="밑줄 (Underline)", variable=underline_var).grid(row=3, column=1, sticky="w", pady=2, padx=5)
        tk.Checkbutton(frame, text="취소선 (Strike)", variable=strike_var).grid(row=3, column=2, sticky="w", pady=2, padx=5)

        def apply_font_settings():
            font_family = font_cb.get()
            font_size = int(size_cb.get())
            
            font_style = tkfont.Font(
                family=font_family,
                size=font_size,
                weight="bold" if bold_var.get() else "normal",
                slant="italic" if italic_var.get() else "roman",
                underline=1 if underline_var.get() else 0,
                overstrike=1 if strike_var.get() else 0
            )

            if has_selection:
                # [상황 1] 사용자가 글자를 선택(드래그)한 경우 -> 고유 태그로 해당 부분만 변경
                self.tag_id += 1
                unique_tag_name = f"style_{self.tag_id}"
                self.text_area.tag_configure(unique_tag_name, font=font_style)
                self.text_area.tag_add(unique_tag_name, start_idx, end_idx)
            else:
                # [상황 2] 선택 영역이 없는 경우 -> 현재 커서 위치 포함 전체 기본 서식 자체를 변경
                # 이 시점부터 작성하는 행과 아래 행들은 바뀐 기본 폰트를 따라갑니다.
                self.current_font_family = font_family
                self.current_font_size = font_size
                self.is_bold = bold_var.get()
                self.is_italic = italic_var.get()
                self.is_underline = underline_var.get()
                self.is_strike = strike_var.get()
                
                self.update_base_font()
            
            font_window.destroy()

        # 4. 확인 / 취소 버튼
        btn_frame = tk.Frame(frame)
        btn_frame.grid(row=4, column=0, columnspan=3, pady=25)

        ok_btn = tk.Button(btn_frame, text="확인", width=10, command=apply_font_settings)
        ok_btn.pack(side=tk.LEFT, padx=7)
        
        cancel_btn = tk.Button(btn_frame, text="취소", width=10, command=font_window.destroy)
        cancel_btn.pack(side=tk.LEFT, padx=7)

    # --- 파일 제어 기능들 ---
    def new_file(self):
        self.root.title("제목 없음 - 파이썬 메모장")
        self.file_path = None
        self.text_area.delete(1.0, tk.END)

    def open_file(self):
        file = filedialog.askopenfile(defaultextension=".txt", 
                                      filetypes=[("텍스트 문서 (*.txt)", "*.txt"), ("모든 파일", "*.*")])
        if file is None:
            return
        self.file_path = file.name
        self.root.title(f"{os.path.basename(self.file_path)} - 파이썬 메모장")
        self.text_area.delete(1.0, tk.END)
        self.text_area.insert(1.0, file.read())
        file.close()

    def save_file(self):
        if self.file_path is None:
            self.save_as_file()
        else:
            try:
                with open(self.file_path, "w", encoding="utf-8") as file:
                    file.write(self.text_area.get(1.0, tk.END))
            except Exception as e:
                messagebox.showerror("오류", f"파일을 저장할 수 없습니다:\n{e}")

    def save_as_file(self):
        file_path = filedialog.asksaveasfilename(initialfile="제목 없음.txt",
                                                defaultextension=".txt",
                                                filetypes=[("텍스트 문서 (*.txt)", "*.txt"), ("모든 파일", "*.*")])
        if not file_path:
            return
        self.file_path = file_path
        self.root.title(f"{os.path.basename(self.file_path)} - 파이썬 메모장")
        self.save_file()

    def quit_application(self):
        self.root.quit()

if __name__ == "__main__":
    root = tk.Tk()
    notepad = Notepad(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import colorchooser
from tkinter import messagebox
from tkinter import ttk
import tkinter.font as tkfont

class PaintApp:
    def __init__(self, root):
        self.root = root
        self.root.title("파이썬 스마트 영역 편집 그림판 (드래그 글자 넣기)")
        self.root.geometry("1150x700")
        
        # --- 초기 설정 변수들 ---
        self.pen_color = "black"       
        self.eraser_color = "white"    
        self.pen_size = 5              
        self.current_mode = "pen"      # pen, eraser, line, rect, oval, select
        self.is_text_placement_mode = False  # ⭐ 글자 넣기용 드래그 대기 상태 플래그
        
        # 마우스 좌표 제어 변수
        self.start_x = None
        self.start_y = None
        self.old_x = None
        self.old_y = None
        
        # 임시 가이드선 및 선택 영역 변수
        self.temp_shape_id = None
        self.selection_box_id = None
        self.selected_items = []       
        self.selected_area = None      

        # --- GUI 레이아웃 구성 ---
        self.controls_frame = tk.Frame(self.root, bg="#e0e0e0", padx=5, pady=5)
        self.controls_frame.pack(side=tk.TOP, fill=tk.X)

        # 기본 도구
        self.color_btn = tk.Button(self.controls_frame, text="🎨 색상", command=self.choose_color)
        self.color_btn.pack(side=tk.LEFT, padx=3)
        self.clear_btn = tk.Button(self.controls_frame, text="🗑️ 초기화", command=self.clear_canvas)
        self.clear_btn.pack(side=tk.LEFT, padx=3)

        tk.Label(self.controls_frame, text=" | ", bg="#e0e0e0", fg="#999999").pack(side=tk.LEFT)

        # 모드 버튼들
        self.pen_btn = tk.Button(self.controls_frame, text="✏️ 펜", bg="#b3e5fc", command=lambda: self.set_mode("pen"))
        self.pen_btn.pack(side=tk.LEFT, padx=2)
        self.eraser_btn = tk.Button(self.controls_frame, text="🧽 지우개", command=lambda: self.set_mode("eraser"))
        self.eraser_btn.pack(side=tk.LEFT, padx=2)
        self.line_btn = tk.Button(self.controls_frame, text="📏 직선", command=lambda: self.set_mode("line"))
        self.line_btn.pack(side=tk.LEFT, padx=2)
        self.rect_btn = tk.Button(self.controls_frame, text="⬜ 사각형", command=lambda: self.set_mode("rect"))
        self.rect_btn.pack(side=tk.LEFT, padx=2)
        self.oval_btn = tk.Button(self.controls_frame, text="⭕ 원", command=lambda: self.set_mode("oval"))
        self.oval_btn.pack(side=tk.LEFT, padx=2)

        tk.Label(self.controls_frame, text=" | ", bg="#e0e0e0", fg="#999999").pack(side=tk.LEFT)

        # 영역 선택 모드 버튼
        self.select_btn = tk.Button(self.controls_frame, text="🔍 영역 선택", command=lambda: self.set_mode("select"))
        self.select_btn.pack(side=tk.LEFT, padx=2)

        # [변형 조작 및 텍스트 도구함]
        self.transform_frame = tk.Frame(self.controls_frame, bg="#d0d0d0", padx=2, pady=2)
        self.transform_frame.pack(side=tk.LEFT, padx=5)
        
        tk.Button(self.transform_frame, text="➕ 확대", command=lambda: self.transform_selected(1.1)).pack(side=tk.LEFT, padx=1)
        tk.Button(self.transform_frame, text="➖ 축소", command=lambda: self.transform_selected(0.9)).pack(side=tk.LEFT, padx=1)
        tk.Button(self.transform_frame, text="↪️ 시계 90도", command=lambda: self.rotate_selected(90)).pack(side=tk.LEFT, padx=1)
        tk.Button(self.transform_frame, text="↩️ 반시계 90도", command=lambda: self.rotate_selected(-90)).pack(side=tk.LEFT, padx=1)
        tk.Button(self.transform_frame, text="↔️ 좌우 대칭", command=lambda: self.flip_selected("horizontal")).pack(side=tk.LEFT, padx=1)
        tk.Button(self.transform_frame, text="↕️ 상하 대칭", command=lambda: self.flip_selected("vertical")).pack(side=tk.LEFT, padx=1)
        
        # 💬 [업그레이드] 글자 넣기 버튼 트리거 변경
        self.text_btn = tk.Button(self.transform_frame, text="💬 글자 넣기", command=self.prepare_text_placement)
        self.text_btn.pack(side=tk.LEFT, padx=3)

        # 두께 입력창
        tk.Label(self.controls_frame, text=" 두께:", bg="#e0e0e0").pack(side=tk.LEFT, padx=2)
        self.size_entry = tk.Entry(self.controls_frame, width=3, justify=tk.CENTER)
        self.size_entry.insert(0, str(self.pen_size))
        self.size_entry.pack(side=tk.LEFT, padx=2)
        self.apply_btn = tk.Button(self.controls_frame, text="OK", command=self.apply_new_size)
        self.apply_btn.pack(side=tk.LEFT, padx=2)

        # 상태 표시
        self.status_label = tk.Label(self.controls_frame, text="모드: 일반 펜", fg="blue", bg="#e0e0e0")
        self.status_label.pack(side=tk.RIGHT, padx=10)

        # 2. 메인 캔버스 영역
        self.canvas = tk.Canvas(self.root, bg="white", highlightthickness=0)
        self.canvas.pack(fill=tk.BOTH, expand=True)

        # 마우스 이벤트 바인딩
        self.canvas.bind("<Button-1>", self.on_button_press)     
        self.canvas.bind("<B1-Motion>", self.on_move_press)       
        self.canvas.bind("<ButtonRelease-1>", self.on_button_release) 

        # 방향키 바인딩 (이동)
        self.root.bind("<Up>", lambda e: self.move_selected(0, -10))
        self.root.bind("<Down>", lambda e: self.move_selected(0, 10))
        self.root.bind("<Left>", lambda e: self.move_selected(-10, 0))
        self.root.bind("<Right>", lambda e: self.move_selected(10, 0))

    def apply_new_size(self):
        try:
            size = int(self.size_entry.get())
            if 1 <= size <= 99: self.pen_size = size
        except ValueError:
            self.size_entry.delete(0, tk.END)
            self.size_entry.insert(0, str(self.pen_size))

    def set_mode(self, mode):
        self.current_mode = mode
        self.is_text_placement_mode = False # 다른 모드로 가면 글자 넣기 상태 해제
        self.clear_selection_box()
        
        for btn in [self.pen_btn, self.eraser_btn, self.line_btn, self.rect_btn, self.oval_btn, self.select_btn, self.text_btn]:
            btn.config(bg=self.root.cget("bg"))
            
        if mode == "pen": self.pen_btn.config(bg="#b3e5fc")
        elif mode == "eraser": self.eraser_btn.config(bg="#b3e5fc")
        elif mode == "line": self.line_btn.config(bg="#b3e5fc")
        elif mode == "rect": self.rect_btn.config(bg="#b3e5fc")
        elif mode == "oval": self.oval_btn.config(bg="#b3e5fc")
        elif mode == "select": self.select_btn.config(bg="#ffecb3")
        
        self.update_status()

    def prepare_text_placement(self):
        """⭐ 글자 넣기 버튼을 눌렀을 때 실행되는 함수 (드래그 대기 모드 활성화)"""
        self.clear_selection_box()
        self.current_mode = "select"  # 내부 메커니즘은 '영역 선택'과 동일하게 설정
        self.is_text_placement_mode = True # 다만 '글씨 넣기용' 드래그라는 플래그 활성화
        
        # 버튼 하이라이트 시각화
        for btn in [self.pen_btn, self.eraser_btn, self.line_btn, self.rect_btn, self.oval_btn, self.select_btn]:
            btn.config(bg=self.root.cget("bg"))
        self.text_btn.config(bg="#d1c4e9") # 보라색 테마 하이라이트
        
        self.status_label.config(text="모드: 글자 영역 드래그 대기 중...", fg="purple")

    def clear_selection_box(self):
        if self.selection_box_id:
            self.canvas.delete(self.selection_box_id)
            self.selection_box_id = None
        self.selected_items = []
        self.selected_area = None

    def on_button_press(self, event):
        self.apply_new_size()
        self.start_x = event.x
        self.start_y = event.y
        self.old_x = event.x
        self.old_y = event.y

        if self.current_mode == "select" and not self.is_text_placement_mode:
            if self.selected_area:
                x1, y1, x2, y2 = self.selected_area
                if not (min(x1, x2) <= event.x <= max(x1, x2) and min(y1, y2) <= event.y <= max(y1, y2)):
                    self.clear_selection_box()
            else:
                self.clear_selection_box()

    def on_move_press(self, event):
        current_color = self.eraser_color if self.current_mode == "eraser" else self.pen_color

        if self.current_mode in ["pen", "eraser"]:
            self.canvas.create_line(self.old_x, self.old_y, event.x, event.y, width=self.pen_size, fill=current_color, capstyle=tk.ROUND, smooth=tk.TRUE)
            self.old_x = event.x
            self.old_y = event.y
            
        elif self.current_mode == "select":
            # 일반 영역선택과 글자 넣기 대기 모드 모두 보라/파란색 가이드라인 점선 박스 표시
            if self.temp_shape_id: self.canvas.delete(self.temp_shape_id)
            outline_color = "purple" if self.is_text_placement_mode else "blue"
            self.temp_shape_id = self.canvas.create_rectangle(self.start_x, self.start_y, event.x, event.y, outline=outline_color, dash=(4, 4))
            
        else:
            if self.temp_shape_id: self.canvas.delete(self.temp_shape_id)
            if self.current_mode == "line":
                self.temp_shape_id = self.canvas.create_line(self.start_x, self.start_y, event.x, event.y, width=self.pen_size, fill=current_color)
            elif self.current_mode == "rect":
                self.temp_shape_id = self.canvas.create_rectangle(self.start_x, self.start_y, event.x, event.y, width=self.pen_size, outline=current_color)
            elif self.current_mode == "oval":
                self.temp_shape_id = self.canvas.create_oval(self.start_x, self.start_y, event.x, event.y, width=self.pen_size, outline=current_color)

    def on_button_release(self, event):
        current_color = self.eraser_color if self.current_mode == "eraser" else self.pen_color
        
        if self.temp_shape_id:
            self.canvas.delete(self.temp_shape_id)
            self.temp_shape_id = None

        if self.current_mode == "select":
            x1, y1, x2, y2 = self.start_x, self.start_y, event.x, event.y
            if abs(x1 - x2) > 5 and abs(y1 - y2) > 5:
                # 빨간 점선 상자 확정 출력
                self.selection_box_id = self.canvas.create_rectangle(x1, y1, x2, y2, outline="red", dash=(4, 4), width=2)
                self.selected_area = (x1, y1, x2, y2)
                
                # 내부 오브젝트 데이터 수집
                self.selected_items = list(self.canvas.find_enclosed(min(x1, x2), min(y1, y2), max(x1, x2), max(y1, y2)))
                overlapping = self.canvas.find_overlapping(min(x1, x2), min(y1, y2), max(x1, x2), max(y1, y2))
                self.selected_items = list(set(self.selected_items) | set(overlapping))
                if self.selection_box_id in self.selected_items:
                    self.selected_items.remove(self.selection_box_id)
                
                # ⭐ 만약 글자 넣기 버튼을 누르고 드래그를 끝낸 시점이라면 곧바로 서식 창 호출!
                if self.is_text_placement_mode:
                    self.open_text_format_window()
        
        elif self.current_mode == "line":
            self.canvas.create_line(self.start_x, self.start_y, event.x, event.y, width=self.pen_size, fill=current_color)
        elif self.current_mode == "rect":
            self.canvas.create_rectangle(self.start_x, self.start_y, event.x, event.y, width=self.pen_size, outline=current_color)
        elif self.current_mode == "oval":
            self.canvas.create_oval(self.start_x, self.start_y, event.x, event.y, width=self.pen_size, outline=current_color)

        self.start_x = self.start_y = self.old_x = self.old_y = None

    # --- 변형 조작 제어 함수부 ---
    def move_selected(self, dx, dy):
        if not self.selected_items or not self.selected_area: return
        for item in self.selected_items:
            self.canvas.move(item, dx, dy)
        if self.selection_box_id:
            self.canvas.move(self.selection_box_id, dx, dy)
        x1, y1, x2, y2 = self.selected_area
        self.selected_area = (x1 + dx, y1 + dy, x2 + dx, y2 + dy)

    def transform_selected(self, scale_factor):
        if not self.selected_items or not self.selected_area: return
        x1, y1, x2, y2 = self.selected_area
        cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
        
        for item in self.selected_items:
            self.canvas.scale(item, cx, cy, scale_factor, scale_factor)
            
        if self.selection_box_id:
            self.canvas.scale(self.selection_box_id, cx, cy, scale_factor, scale_factor)
            
        nx1 = cx + (x1 - cx) * scale_factor
        ny1 = cy + (y1 - cy) * scale_factor
        nx2 = cx + (x2 - cx) * scale_factor
        ny2 = cy + (y2 - cy) * scale_factor
        self.selected_area = (nx1, ny1, nx2, ny2)

    def rotate_selected(self, angle_degrees):
        if not self.selected_items or not self.selected_area: return
        x1, y1, x2, y2 = self.selected_area
        cx, cy = (x1 + x2) / 2, (y1 + y2) / 2 

        for item in self.selected_items:
            coords = self.canvas.coords(item)
            new_coords = []
            
            if self.canvas.type(item) == "text":
                x, y = coords[0], coords[1]
                if angle_degrees == 90:
                    rx, ry = cx - (y - cy), cy + (x - cx)
                else:
                    rx, ry = cx + (y - cy), cy - (x - cx)
                self.canvas.coords(item, rx, ry)
                continue

            for i in range(0, len(coords), 2):
                x, y = coords[i], coords[i+1]
                if angle_degrees == 90:
                    rx, ry = cx - (y - cy), cy + (x - cx)
                else:
                    rx, ry = cx + (y - cy), cy - (x - cx)
                new_coords.extend([rx, ry])
                
            self.canvas.coords(item, *new_coords)

    def flip_selected(self, direction):
        if not self.selected_items or not self.selected_area: return
        x1, y1, x2, y2 = self.selected_area
        cx, cy = (x1 + x2) / 2, (y1 + y2) / 2

        for item in self.selected_items:
            coords = self.canvas.coords(item)
            new_coords = []
            
            if self.canvas.type(item) == "text":
                x, y = coords[0], coords[1]
                rx = (2 * cx - x) if direction == "horizontal" else x
                ry = y if direction == "horizontal" else (2 * cy - y)
                self.canvas.coords(item, rx, ry)
                continue
                
            for i in range(0, len(coords), 2):
                x, y = coords[i], coords[i+1]
                rx = (2 * cx - x) if direction == "horizontal" else x
                ry = y if direction == "horizontal" else (2 * cy - y)
                new_coords.extend([rx, ry])
                
            self.canvas.coords(item, *new_coords)

    def open_text_format_window(self):
        """글꼴/스타일 지정 팝업창 및 배치 확정"""
        # 글자 삽입용 드래그 기능이 성공적으로 완료되었으므로 상태 초기화
        self.is_text_placement_mode = False

        text_window = tk.Toplevel(self.root)
        text_window.title("텍스트 삽입 및 서식 설정")
        text_window.geometry("400x380")
        text_window.resizable(False, False)
        text_window.transient(self.root)
        text_window.grab_set()

        frame = tk.Frame(text_window, padx=15, pady=15)
        frame.pack(fill=tk.BOTH, expand=True)

        # 1. 텍스트 내용 입력 필드
        tk.Label(frame, text="내용 입력:").grid(row=0, column=0, sticky="w", pady=3)
        text_entry = tk.Entry(frame, width=32)
        text_entry.grid(row=0, column=1, columnspan=2, pady=3, sticky="w")
        text_entry.focus_set()

        # 2. 글꼴 드롭다운
        tk.Label(frame, text="글꼴 선택:").grid(row=1, column=0, sticky="w", pady=5)
        filtered_fonts = [f for f in sorted(list(tkfont.families())) if not f.startswith('@')]
        font_cb = ttk.Combobox(frame, values=filtered_fonts, width=22, state="readonly")
        font_cb.set("맑은 고딕")
        font_cb.grid(row=1, column=1, columnspan=2, pady=5, sticky="w")

        # 3. 글자 크기 드롭다운
        tk.Label(frame, text="글자 크기:").grid(row=2, column=0, sticky="w", pady=5)
        sizes = [10, 12, 14, 16, 18, 20, 24, 28, 32, 36, 40, 48, 64, 72]
        size_cb = ttk.Combobox(frame, values=sizes, width=22, state="readonly")
        size_cb.set(24) 
        size_cb.grid(row=2, column=1, columnspan=2, pady=5, sticky="w")

        # 4. 서식 스타일 효과 체크박스
        tk.Label(frame, text="스타일 효과:").grid(row=3, column=0, sticky="nw", pady=8)
        
        bold_var = tk.BooleanVar(value=False)
        italic_var = tk.BooleanVar(value=False)
        underline_var = tk.BooleanVar(value=False)
        strike_var = tk.BooleanVar(value=False)

        tk.Checkbutton(frame, text="굵게 (Bold)", variable=bold_var).grid(row=3, column=1, sticky="w", pady=2)
        tk.Checkbutton(frame, text="기울임 (Italic)", variable=italic_var).grid(row=3, column=2, sticky="w", pady=2)
        tk.Checkbutton(frame, text="밑줄 (Underline)", variable=underline_var).grid(row=4, column=1, sticky="w", pady=2)
        tk.Checkbutton(frame, text="취소선 (Strike)", variable=strike_var).grid(row=4, column=2, sticky="w", pady=2)

        def create_formatted_text():
            user_text = text_entry.get().strip()
            if not user_text:
                messagebox.showwarning("주의", "글자 내용을 입력해 주세요.")
                return

            x1, y1, x2, y2 = self.selected_area
            cx, cy = (x1 + x2) / 2, (y1 + y2) / 2 

            font_family = font_cb.get()
            font_size = int(size_cb.get())
            
            custom_font = tkfont.Font(
                family=font_family,
                size=font_size,
                weight="bold" if bold_var.get() else "normal",
                slant="italic" if italic_var.get() else "roman",
                underline=1 if underline_var.get() else 0,
                overstrike=1 if strike_var.get() else 0
            )

            text_id = self.canvas.create_text(
                cx, cy, 
                text=user_text, 
                fill=self.pen_color, 
                font=custom_font
            )
            
            self.selected_items.append(text_id)
            self.text_btn.config(bg=self.root.cget("bg")) # 글자 삽입 완료 시 버튼 색상 원복
            self.update_status()
            text_window.destroy()

        def cancel_text_placement():
            self.clear_selection_box()
            self.text_btn.config(bg=self.root.cget("bg"))
            self.update_status()
            text_window.destroy()

        # 5. 하단 버튼 배치
        btn_frame = tk.Frame(frame)
        btn_frame.grid(row=5, column=0, columnspan=3, pady=25)

        tk.Button(btn_frame, text="확인", width=12, command=create_formatted_text).pack(side=tk.LEFT, padx=5)
        tk.Button(btn_frame, text="취소", width=12, command=cancel_text_placement).pack(side=tk.LEFT, padx=5)

        # 팝업 엑스(X) 단추로 껐을 때를 위한 예외 처리 핸들러 바인딩
        text_window.protocol("WM_DELETE_WINDOW", cancel_text_placement)

    def choose_color(self):
        color = colorchooser.askcolor(color=self.pen_color)[1]
        if color: self.pen_color = color

    def clear_canvas(self):
        if messagebox.askyesno("확인", "도화지를 초기화하시겠습니까?"):
            self.canvas.delete("all")
            self.clear_selection_box()

    def update_status(self):
        mode_names = {"pen": "일반 펜", "eraser": "지우개", "line": "직선", "rect": "사각형", "oval": "원", "select": "영역 선택 모드"}
        self.status_label.config(text=f"모드: {mode_names[self.current_mode]}", fg="blue")

if __name__ == "__main__":
    root = tk.Tk()
    app = PaintApp(root)
    root.mainloop()
```
