```python
import matplotlib.pylab as plt
import numpy as np
from sklearn import linear_model

# 선형 회귀 모델을 생성한다. 
reg = linear_model.LinearRegression()

# 데이터는 파이썬의 리스트로 만들어도 되고 아니면 넘파이의 배열로 만들어도 됨
X = [[174], [152], [138], [128], [186]]     # 학습 예제 (키)
y = [71, 55, 46, 38, 88]                # 정답 (몸무게)

reg.fit(X, y)                        # 학습 함수

print(f"키 193cm일 때 예측 몸무게: {reg.predict([[193]])}") # 예측 함수

# --- 여기서부터 그래프 그리는 코드 추가 ---

# 1. 실제 데이터(X, y)를 산점도(점 그래프)로 그립니다.
plt.scatter(X, y, color='blue', label='Actual Data')

# 2. 선형 회귀선(Trend Line)을 그리기 위한 가상의 X축 데이터 생성 (120부터 200까지)
X_line = np.linspace(120, 200, 100).reshape(-1, 1)
y_line = reg.predict(X_line)  # 해당 가상 데이터들에 대한 모델의 예측값 계산

# 3. 모델이 예측한 회귀선을 빨간색 실선으로 그립니다.
plt.plot(X_line, y_line, color='red', linewidth=2, label='Regression Line')

# 4. 방금 예측했던 193cm 데이터도 그래프에 초록색 별 모양으로 표시해 볼게요.
plt.scatter([[193]], reg.predict([[193]]), color='green', marker='*', s=200, label='Prediction (193cm)')

# 5. 그래프 꾸미기 (제목, 축 이름, 범례)
plt.title('Linear Regression Example')
plt.xlabel('X (e.g., Height)')
plt.ylabel('y (e.g., Weight)')
plt.legend()  # 범례 표시
plt.grid(True) # 모눈종이 격자 표시

# 6. 화면에 그래프 띄우기
plt.show()
```
